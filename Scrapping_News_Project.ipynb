```python
# Importing the required libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Fetching the response from the url and saving the content from the response using BeautifulSoup
url = 'https://www.aljazeera.com/xml/rss/all.xml'
response = requests.get(url)
soup = BeautifulSoup(response.content, features = 'xml') 


# Prettify () function will turn a BeautifulSoup parse tree into a structured Unicode string, 
# with a separate line for each tag and each string 
print(soup.prettify())
```

```python
# Finding all the data under 'item' tag using findAll() function
articles = soup.findAll('item')

articles
```

```python
# Declaring an empty list to save our scrapped data
scrapped_data = []

# Appending the scrapped data from our soup to the above declared list
for article in articles:
    dict_items = {}
    dict_items['title'] = article.title.text
    dict_items['published_date'] = article.pubDate.text
    scrapped_data.append(dict_items)
    
scrapped_data
```

```python
# importing the pandas library 
import pandas as pd

# Creating a DataFrame with the scrapped data
Articles = pd.DataFrame(scrapped_data, columns = ['title', 'published_date'])

# Exporting the DataFrame to a csv file
Articles.to_csv('AlJazeera_headlines.csv',index=False, encoding = 'utf-8')
```
